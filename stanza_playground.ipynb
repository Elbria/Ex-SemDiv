```python
import stanza
from stanza.server import CoreNLPClient
import string
import re
#stanza.install_corenlp()
data = 'annotations/ced_en_pt_with_labels_itermax_pos'
#data = 'annotations/mqm_generalMT2022_en_de_itermax_pos'
```

```python
punctuation_types = [punc for punc in string.punctuation]
excluded_types = punctuation_types + ['-LRB-', '-RRB-', 'DT', 'IN', '``', 'HYPH', 'CC', 'TO' ]

def parse_text(texts):
    anns = []
    with CoreNLPClient(
            annotators=[ 'parse, pos'],
            timeout=30000,
            memory='6G') as client:        
        for line in texts:        
            anns.append(client.annotate(line, properties={'tokenize.whitespace': 'true'}))
    return anns

def pos_parse_text(texts):
    anns = []
    with CoreNLPClient(
            properties='spanish', 
            annotators=[ 'pos'],
            timeout=30000,
            memory='6G'
            ) as client:        
        for line in texts:        
            anns.append(client.annotate(line, properties={'tokenize.whitespace': 'true'}))
    return anns

    
class ConstituentPhrases:
    def __init__(self, parse_tree, phrases, types, parent, excluded_types):   
        self.value = parse_tree.value
        self.score = parse_tree.score
        self.children = [ConstituentPhrases(child, phrases, types, self.value , excluded_types) for child in parse_tree.child]
        self.phrases = phrases
        self.types = types
        self.excluded_types = excluded_types
        self.parent = parent
            
    def __str__(self):
        if self.children:
            self.types.add(self.value)
            if self.value not in self.excluded_types:
                self.phrases.add(f"{' '.join(str(child) for child in self.children)}")
            return f"{' '.join(str(child) for child in self.children)}"
        if self.value not in self.excluded_types and self.parent not in self.excluded_types:
            self.phrases.add(f"{self.value}")
        return f"{self.value}"
```

```python
texts = []
with open(data, 'r') as file_:
    lines = file_.readlines()
    for id_, line in enumerate(lines):
        #if id_ > 20:
        #    continue
        line = line.split('\t')
        texts.append(line[1])
        
```

```python
if not pos:
    extracted_phrases_all = []
    for id_, ann in enumerate(anns):
        print(id_)
        parse_tree = ConstituentPhrases(ann.sentence[0].parseTree.child[0], phrases=set(), types=set(), parent='S', excluded_types=excluded_types)
        parse_tree.__str__()
        extracted_phrases = parse_tree.phrases
        extracted_phrases_all.append(extracted_phrases)
        #extracted_types = parse_tree.types
```

```python
if not pos:
    character_indices_all = []
    for j in range(len(texts)):

        input_string = texts[j]
        syntactic_phrases = list(extracted_phrases_all[j])
        character_indices = []

        for phrase in syntactic_phrases:
            index_ = input_string.index(phrase)
            character_indices.append(f'{index_}:{index_+len(phrase)}')
        character_indices_all.append(character_indices)
```

```python
if not pos:
    with open(f'{data}_parse', 'w') as output_:
        for x, char_ind in enumerate(character_indices_all):
            line = lines[x].rstrip().split('\t')
            new_line = line + [' '.join(char_ind)]
            output_.write('\t'.join(new_line) + '\n')
    output_.close()
```
