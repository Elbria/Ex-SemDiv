```python
import sys

sys.path.append('./divergentmBERT')
import pandas as pd
import numpy as np
import shap
import seaborn as sns
import os
import argparse
from divergentscorer import DivergentmBERTScorer
import torch
device = torch.device("cuda" if torch.cuda.is_available() and not True else "cpu")
n_gpu = torch.cuda.device_count()
device = device
```

```python
class DivergentmBERT():
    def __init__(self, model, tokenizer, do_lower_case, device):
        self.scorer = DivergentmBERTScorer(model, tokenizer, do_lower_case, device)

    def __call__(self, text_a, text_b):
        return self.scorer.compute_divergentscore(text_a, text_b)
```

```python
model = DivergentmBERT(model='/fs/clip-divergences/xling-SemDiv/trained_bert/from_WikiMatrix.en-fr.tsv.filtered_sample_50000.moses.seed/contrastive_divergence_ranking/rdpg',
                                  tokenizer='bert-base-multilingual-cased',
                                  do_lower_case=False,
                                  device=device)

#model_el = DivergentmBERT(model='/fs/clip-divergences/xling-SemDiv/trained_bert/from_WikiMatrix.el-en.tsv.filtered_sample_50000.moses.seed/contrastive_divergence_ranking/rdpg',
#                                  tokenizer='bert-base-multilingual-cased',
#                                  do_lower_case=False,
#                                  device=device)

model_es = DivergentmBERT(model='/fs/clip-divergences/xling-SemDiv/trained_bert_new/from_WikiMatrix.en-es.tsv.filtered_sample_50000.moses.seed/contrastive_divergence_ranking/rdpg',
                                  tokenizer='bert-base-multilingual-cased',
                                  do_lower_case=False,
                                  device=device)
```

```python
model_de = DivergentmBERT(model='/fs/clip-divergences/xling-SemDiv/trained_bert/from_WikiMatrix.de-en.tsv.filtered_sample_50000.moses.seed/contrastive_divergence_ranking/rdpg',
                                  tokenizer='bert-base-multilingual-cased',
                                  do_lower_case=False,
                                  device=device)
```

```python
main_path = 'flores200_dataset/devtest'
source = 'eng_Latn.devtest'
targets = ['fra_Latn.devtest', 'spa_Latn.devtest', 'deu_Latn.devtest', 'ell_Grek.devtest']

english_texts = []
with open(f"{main_path}/{source}", 'r') as file_:
    english_texts = file_.readlines()
    
french_texts = []
with open(f"{main_path}/{targets[0]}", 'r') as file_:
    french_texts = file_.readlines()

spanish_texts = []
with open(f"{main_path}/{targets[1]}", 'r') as file_:
    spanish_texts = file_.readlines()
    
#german_texts = []
#with open(f"{main_path}/{targets[2]}", 'r') as file_:
#    german_texts = file_.readlines()
    
greek_texts = []
with open(f"{main_path}/{targets[3]}", 'r') as file_:
    greek_texts = file_.readlines()
```

```python
en_fr_preds, en_es_preds, en_de_preds, en_el_preds = [], [], [], []
for i in range(len(english_texts)):
    #if i > 100:
    #    continue
    en_fr_pred = model([english_texts[i]], [french_texts[i]])
    en_es_pred = model_es([english_texts[i]], [spanish_texts[i]])
    #en_de_pred = model([english_texts[i]], [german_texts[i]])
    en_el_pred = model_el([english_texts[i]], [greek_texts[i]])

    
    en_fr_preds.append(en_fr_pred)
    en_es_preds.append(en_es_pred)
    #en_de_preds.append(en_de_pred)
    en_el_preds.append(en_el_pred)

    
    print(f'\nSentence: {i}')
    print(f'French: {en_fr_pred[0][0]}')
    print(f'Spanish: {en_es_pred[0][0]}')
    #print(f'German: {en_de_pred[0][0]}')
    print(f'Greek: {en_el_pred[0][0]}')
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))
sns.distplot(en_fr_preds, kde=False, label='French')
#sns.distplot(en_es_preds)
#sns.distplot(en_de_preds)
sns.distplot(en_el_preds, kde=False, label='Greek')
sns.distplot(en_es_preds, kde=False, label='Spanish')

sns.set_style("darkgrid")

plt.rcParams.update({'font.size': 15})
fig.legend(labels=['French','Greek','Spanish'])
plt.show()
```

```python
main_path = 'flores200_dataset/devtest'


fra_fra, fra_eng = [], []
with open(f'{main_path}/fra_Latn.devtest.divergent', 'r') as fra:
    fra = fra.readlines()
    for line in fra:
        line = line.rstrip().split('\t')
        fra_fra.append(line[1])
        fra_eng.append(line[0])

        
ell_ell, ell_eng = [], []
with open(f'{main_path}/ell_Grek.devtest.divergent', 'r') as ell:
    ell = ell.readlines()
    for line in ell:
        line = line.rstrip().split('\t')
        ell_ell.append(line[1])
        ell_eng.append(line[0])
        

spa_spa, spa_eng = [], []
with open(f'{main_path}/spa_Latn.devtest.divergent', 'r') as spa:
    spa = spa.readlines()
    for line in spa:
        line = line.rstrip().split('\t')
        spa_spa.append(line[1])
        spa_eng.append(line[0])
```

```python
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
```

```python
fra_ell_spa_overlap_count = 0
for text in fra_eng:
    #print(text)
    try:
        spanish = spa_spa[spa_eng.index(text)] 
    except ValueError:
        continue
    
    try:
        
        print(bcolors.OKBLUE + ell_ell[ell_eng.index(text)] + bcolors.ENDC)
        print(f'{text}')
        print(bcolors.FAIL + fra_fra[fra_eng.index(text)] + bcolors.ENDC + '\n')

    except ValueError:
        continue
        
    fra_ell_spa_overlap_count += 1
```
