```python
fra_scores = []
with open('devtest/fra_Latn.devtest.divergent_scores', 'r') as f:
    f = f.readlines()[1:]
    for id_, line in enumerate(f):
        
        score = float(line.rstrip().split('\t')[0])
        if score > 10:
            fra_scores.append(id_)
            
spa_scores = []
with open('devtest/spa_Latn.devtest.divergent_scores', 'r') as f:
    f = f.readlines()[1:]
    for id_, line in enumerate(f):
        
        score = float(line.rstrip().split('\t')[0])
        if score > 10:
            spa_scores.append(id_)
```

```python
z = list(set(fra_scores).intersection(set(spa_scores))) 
```

```python
output = open('devtest/spa_Latn.devtest.equivalents_', 'w')

with open('devtest/spa_Latn.devtest', 'r') as f:
    f = f.readlines()
    for a in z:
        output.write(f[a])
        print(f[a])
output.close()
```
